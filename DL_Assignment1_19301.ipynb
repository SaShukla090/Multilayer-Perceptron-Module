```python
%%time
from sklearn.metrics import accuracy_score
from activation_layer import Activation_Layer
from fc_layer import FC_layer
from Network import Neural_Network_Classifier
from softmax import softmax_layer
```

```python
%%time
learning_rate = [0.00001, 0.0001, 0.001, 0.01, 0.1]
record = {}
for i in learning_rate:
    print(f"Training for learning rate {i}:")
    nn_classifier = Neural_Network_Classifier()
    nn_classifier.add_layer(FC_layer(16,10 , weight_inialization="xavier")) # 1st fully connected hidden layer
    nn_classifier.add_layer(Activation_Layer(A_funct="tanh")) # corresponding activation layer
    nn_classifier.add_layer(FC_layer(10, 26, weight_inialization="xavier")) # output layer (2nd layer)
    nn_classifier.add_layer(softmax_layer(26))# output layer

    loss = nn_classifier.train(X_train, y_train,X_val, y_val, optimizer="adam", epoch=100, learning_rate = i)
    training_loss, val_loss = loss
    record[str(i)] = training_loss




    yhat = np.array(nn_classifier.predict(X_test))
    yhat = en.inverse_transform(yhat)
    print(f"Accuracy on test set: {accuracy_score(yhat,y_test)}")
    
    
```
