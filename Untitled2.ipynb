```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.random.normal(0,1,(10,))
Y = np.random.normal(0,1,(100,))
```

```python
plt.figure(figsize=(10,8))
plt.plot(X)
plt.title("numbers plots")
plt.xlabel("indices + 1")
plt.ylabel("random number from normal distribution")
plt.legend()
plt.savefig
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.title("dsag")
plt.plot(X)
plt.subplot(1,2,2)
plt.title("SDfa")
plt.plot(Y)
```
